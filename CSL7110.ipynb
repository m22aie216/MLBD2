```python
# Step 4: Define MinHash function
def minhash_signature(binary_matrix, num_hashes=100):
    num_users, num_movies = binary_matrix.shape
    signatures = np.full((num_users, num_hashes), np.inf)
    
    for i in range(num_hashes):
        perm = np.random.permutation(num_movies)
        for user in range(num_users):
            rated = np.where(binary_matrix.iloc[user].values == 1)[0]
            if len(rated) > 0:
                signatures[user, i] = min(perm[rated])
    return signatures
```

```python
import random
from collections import defaultdict

# Step 9: Locality Sensitive Hashing (LSH)
def lsh(signature_matrix, num_bands=20, rows_per_band=5):
    """
    signature_matrix: numpy array (users x num_hashes)
    num_bands: number of bands
    rows_per_band: rows per band (num_bands * rows_per_band = num_hashes)
    """
    num_users, num_hashes = signature_matrix.shape
    assert num_bands * rows_per_band == num_hashes, "Bands * rows_per_band must equal num_hashes"
    
    buckets = defaultdict(list)
    
    for band in range(num_bands):
        start = band * rows_per_band
        end = start + rows_per_band
        
        for user in range(num_users):
            band_signature = tuple(signature_matrix[user, start:end])
            bucket_key = hash(band_signature)
            buckets[bucket_key].append(user)
    
    # Candidate pairs: users that fall into the same bucket
    candidate_pairs = set()
    for bucket_users in buckets.values():
        if len(bucket_users) > 1:
            for i in range(len(bucket_users)):
                for j in range(i+1, len(bucket_users)):
                    candidate_pairs.add((bucket_users[i], bucket_users[j]))
                    
    return candidate_pairs

# Step 10: Run LSH
candidate_pairs = lsh(signature_matrix, num_bands=20, rows_per_band=5)
print("Number of candidate pairs found:", len(candidate_pairs))

# Step 11: Evaluate candidate pairs
results = []
for u1, u2 in list(candidate_pairs)[:10]:  # show first 10 pairs
    est_sim = jaccard_similarity(signature_matrix, u1, u2)
    true_sim = true_jaccard(binary_matrix, u1, u2)
    results.append((u1, u2, est_sim, true_sim))

results_df = pd.DataFrame(results, columns=["User1","User2","EstimatedSim","TrueSim"])
print(results_df)

# Step 12: Visualization of candidate similarities
plt.scatter(results_df["EstimatedSim"], results_df["TrueSim"], c="blue")
plt.xlabel("Estimated Jaccard (MinHash)")
plt.ylabel("True Jaccard")
plt.title("LSH Candidate Pair Similarities")
plt.show()
```

```python
import pandas as pd
import numpy as np
from itertools import combinations

# Load MovieLens 100k dataset
df = pd.read_csv("ml-100k/u.data", sep="\t", names=["user_id","movie_id","rating","timestamp"])

# Binary user-movie matrix
binary_matrix = df.groupby(["user_id","movie_id"]).size().unstack(fill_value=0)
binary_matrix = (binary_matrix > 0).astype(int)

num_users, num_movies = binary_matrix.shape

# Exact Jaccard similarity
def true_jaccard(binary_matrix, u1, u2):
    set1 = set(np.where(binary_matrix.iloc[u1].values == 1)[0])
    set2 = set(np.where(binary_matrix.iloc[u2].values == 1)[0])
    return len(set1 & set2) / len(set1 | set2) if len(set1 | set2) > 0 else 0

# Compute exact pairs with similarity >= 0.5
exact_pairs = []
for u1, u2 in combinations(range(num_users), 2):
    sim = true_jaccard(binary_matrix, u1, u2)
    if sim >= 0.5:
        exact_pairs.append((u1, u2, sim))
exact_set = {(u1,u2) for u1,u2,_ in exact_pairs}

# Min-Hash signatures
def minhash_signature(binary_matrix, num_hashes=100):
    num_users, num_movies = binary_matrix.shape
    signatures = np.full((num_users, num_hashes), np.inf)
    
    for i in range(num_hashes):
        perm = np.random.permutation(num_movies)
        for user in range(num_users):
            rated = np.where(binary_matrix.iloc[user].values == 1)[0]
            if len(rated) > 0:
                signatures[user, i] = min(perm[rated])
    return signatures

def jaccard_similarity(sig_matrix, u1, u2):
    return np.mean(sig_matrix[u1] == sig_matrix[u2])

# Evaluate one run
def evaluate_run(binary_matrix, num_hashes=100):
    sig_matrix = minhash_signature(binary_matrix, num_hashes)
    approx_pairs = []
    for u1, u2 in combinations(range(num_users), 2):
        sim = jaccard_similarity(sig_matrix, u1, u2)
        if sim >= 0.5:
            approx_pairs.append((u1, u2, sim))
    approx_set = {(u1,u2) for u1,u2,_ in approx_pairs}
    
    false_positives = len(approx_set - exact_set)
    false_negatives = len(exact_set - approx_set)
    return false_positives, false_negatives

# Run for 50, 100, 200 hash functions
hash_sizes = [50, 100, 200]
results = {}

for h in hash_sizes:
    fp_list, fn_list = [], []
    for run in range(5):
        fp, fn = evaluate_run(binary_matrix, num_hashes=h)
        fp_list.append(fp)
        fn_list.append(fn)
        print(f"Hash={h}, Run {run+1}: FP={fp}, FN={fn}")
    results[h] = {
        "avg_fp": np.mean(fp_list),
        "avg_fn": np.mean(fn_list)
    }

print("\nSummary of averages:")
for h in hash_sizes:
    print(f"{h} hashes -> Avg False Positives={results[h]['avg_fp']}, Avg False Negatives={results[h]['avg_fn']}")
```

```python
from collections import defaultdict
from itertools import combinations

# LSH function
def lsh(signature_matrix, r, b, threshold=0.6):
    """
    signature_matrix: numpy array (users x num_hashes)
    r: rows per band
    b: number of bands
    threshold: similarity cutoff
    """
    num_users, num_hashes = signature_matrix.shape
    assert r * b == num_hashes, "r * b must equal num_hashes"
    
    buckets = defaultdict(list)
    
    # Hash each band
    for band in range(b):
        start = band * r
        end = start + r
        for user in range(num_users):
            band_signature = tuple(signature_matrix[user, start:end])
            bucket_key = hash(band_signature)
            buckets[(band, bucket_key)].append(user)
    
    # Candidate pairs
    candidate_pairs = set()
    for bucket_users in buckets.values():
        if len(bucket_users) > 1:
            for u1, u2 in combinations(bucket_users, 2):
                candidate_pairs.add((u1, u2))
    
    # Filter by estimated similarity
    results = []
    for u1, u2 in candidate_pairs:
        sim = jaccard_similarity(signature_matrix, u1, u2)
        if sim >= threshold:
            results.append((u1, u2, sim))
    return results

# Experiment runner
def run_lsh_experiment(binary_matrix, num_hashes, r, b, threshold=0.6, runs=5):
    fp_list, fn_list = [], []
    for run in range(runs):
        sig_matrix = minhash_signature(binary_matrix, num_hashes)
        approx_pairs = lsh(sig_matrix, r, b, threshold)
        approx_set = {(u1,u2) for u1,u2,_ in approx_pairs}
        
        # Exact set for threshold
        exact_pairs = []
        for u1, u2 in combinations(range(num_users), 2):
            sim = true_jaccard(binary_matrix, u1, u2)
            if sim >= threshold:
                exact_pairs.append((u1, u2, sim))
        exact_set = {(u1,u2) for u1,u2,_ in exact_pairs}
        
        fp = len(approx_set - exact_set)
        fn = len(exact_set - approx_set)
        fp_list.append(fp)
        fn_list.append(fn)
        print(f"Run {run+1}: FP={fp}, FN={fn}")
    
    print(f"\nAverage FP={np.mean(fp_list)}, Average FN={np.mean(fn_list)}")

# Run experiments
print("50 hashes, r=5, b=10")
run_lsh_experiment(binary_matrix, num_hashes=50, r=5, b=10)

print("\n100 hashes, r=5, b=20")
run_lsh_experiment(binary_matrix, num_hashes=100, r=5, b=20)

print("\n200 hashes, r=5, b=40")
run_lsh_experiment(binary_matrix, num_hashes=200, r=5, b=40)

print("\n200 hashes, r=10, b=20")
run_lsh_experiment(binary_matrix, num_hashes=200, r=10, b=20)
```

```python
def run_lsh_experiment(binary_matrix, num_hashes, r, b, threshold=0.6, runs=5):
    fp_list, fn_list = [], []
    for run in range(runs):
        sig_matrix = minhash_signature(binary_matrix, num_hashes)
        approx_pairs = lsh(sig_matrix, r, b, threshold)
        approx_set = {(u1,u2) for u1,u2,_ in approx_pairs}
        
        # Exact set for threshold
        exact_pairs = []
        for u1, u2 in combinations(range(num_users), 2):
            sim = true_jaccard(binary_matrix, u1, u2)
            if sim >= threshold:
                exact_pairs.append((u1, u2, sim))
        exact_set = {(u1,u2) for u1,u2,_ in exact_pairs}
        
        fp = len(approx_set - exact_set)
        fn = len(exact_set - approx_set)
        fp_list.append(fp)
        fn_list.append(fn)
        print(f"Run {run+1}: FP={fp}, FN={fn}")
    
    return np.mean(fp_list), np.mean(fn_list)

# Configurations
configs = [
    (50, 5, 10),
    (100, 5, 20),
    (200, 5, 40),
    (200, 10, 20)
]

# Run experiments for threshold 0.6 and 0.8
results = {}
for num_hashes, r, b in configs:
    print(f"\nConfig: {num_hashes} hashes, r={r}, b={b}, threshold=0.6")
    fp_avg, fn_avg = run_lsh_experiment(binary_matrix, num_hashes, r, b, threshold=0.6)
    results[(num_hashes,r,b,0.6)] = (fp_avg, fn_avg)
    print(f"Average FP={fp_avg}, Average FN={fn_avg}")
    
    print(f"\nConfig: {num_hashes} hashes, r={r}, b={b}, threshold=0.8")
    fp_avg, fn_avg = run_lsh_experiment(binary_matrix, num_hashes, r, b, threshold=0.8)
    results[(num_hashes,r,b,0.8)] = (fp_avg, fn_avg)
    print(f"Average FP={fp_avg}, Average FN={fn_avg}")
```
